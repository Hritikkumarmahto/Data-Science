Write a Python program to display the odd and even elements from the lis


```python
class classOddEven:
  def __init__(self,numbers):
    numbers=numbers
  
  def checkEvenOdd(self,numbers):
    evens=[]
    odds=[]
    for i in numbers:
      if i%2==0:
        evens.append(i)
      else:
        odds.append(i)
    return evens,odds
  
numbers=[1,2,3,4,5,6,6,8,9,3,4,6]
c1=classOddEven(numbers)
res=c1.checkEvenOdd(numbers)
print(res)
```

Write a Python program to interchange the first and last elements in the list

```python
class interChange:
  def __init__(self,numbers):
    self.numbers=numbers
  def changeLastAndFirst(self,numbers):
    temp=0
    size=len(numbers)
    temp=numbers[0]
    numbers[0]=numbers[size-1]
    numbers[size-1]=temp

    return numbers

numbers=[1,2,3,4,5,6,7]
c1=interChange(numbers)
print(c1.changeLastAndFirst(numbers))




```

```python
class duplicateList:
  def __init__(self,numbers,n):
    self.numbers=numbers
    self.n=n
  
  def duplicate(self):
    res=[]
    for j in range(self.n):
      for i in range(len(self.numbers)):
        res.append(self.numbers[i])
    return res
numbers=[1,2,3]
n=6
c1=duplicateList(numbers,n)
res=c1.duplicate()
print(res)



```

Find the smallest element and biggest element in the list 



```python
class FindMinMax:

    def __init__(self, numbers):
        self.numbers = numbers

    def findMax(self):
        maxi = self.numbers[0]

        for i in self.numbers:
            if i > maxi:
                maxi = i

        return maxi

    def findMin(self):
        mini = self.numbers[0]

        for i in self.numbers:
            if i < mini:
                mini = i

        return mini


numbers = []

n = int(input("Enter total number of elements: "))

for i in range(n):
    numbers.append(int(input()))

obj = FindMinMax(numbers)

choice = int(input("1.Min  2.Max : "))

if choice == 1:
    print(obj.findMin())

elif choice == 2:
    print(obj.findMax())
```

Find the repetitive elements in the list


```python
class findRepeat:
  def __init__(self,numbers):
    self.numbers=numbers
    
  def findRepeat(self):
    for i in range(len(self.numbers)-1):
      for j in range(len(i+1,self.numbers)):
        if self.numbers[i] ==self.numbers[j]:
          numbers.pop(j)
    return numbers

numbers=[1,2,3,4,4,4,2,3,4]
c1=findRepeat(numbers)
res=c1.findRepeat()
print(res)
```
