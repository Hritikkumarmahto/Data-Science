## 28. Create list of prime numbers from a list.

```
def prime(n):
    if n<2:return False
    for i in range(2,n):
        if n%i==0:return False
    return True
li=[2,3,4,5,6,7]
print([i for i in li if prime(i)])
```

## 36. Alternate positive and negative numbers.

```
li=[1,2,-3,-4,5,-6]
pos=[i for i in li if i>0]
neg=[i for i in li if i<0]
res=[]
for i in range(min(len(pos),len(neg))):
    res.append(pos[i]);res.append(neg[i])
print(res)
```

## 37. Find pairs whose sum equals target.

```
li=[2,4,3,5,7,8,9]
t=7
for i in range(len(li)):
    for j in range(i+1,len(li)):
        if li[i]+li[j]==t:
            print(li[i],li[j])
```

## 41. Extract digits from strings.

```
li=['abc123','45def','gh6']
res=[]
for s in li:
    d=''
    for ch in s:
        if ch.isdigit():d+=ch
    res.append(d)
print(res)
```

## 42. Find anagram groups.

```
words=['eat','tea','tan','ate','nat','bat']
g={}
for w in words:
    k=''.join(sorted(w))
    if k not in g:g[k]=[]
    g[k].append(w)
print(list(g.values()))
```
