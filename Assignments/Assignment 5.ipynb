1. Filter vowels and consonants from the string "How are you sir".

```python
s="How are you sir"
v="aeiouAEIOU"
print("Vowels:","".join(ch for ch in s if ch in v))
print("Consonants:","".join(ch for ch in s if ch.isalpha() and ch not in v))
```

2. Count vowels and consonants in the string "How are you sir".

```python
s="How are you sir"
v="aeiouAEIOU"
print("Vowels:",sum(ch in v for ch in s))
print("Consonants:",sum(ch.isalpha() and ch not in v for ch in s))
```

13. Count spaces, digits, alphabets, and special characters in "Python 3.9 is awesome!!".

```python
s="Python 3.9 is awesome!!"
spaces=sum(c.isspace() for c in s)
digits=sum(c.isdigit() for c in s)
alpha=sum(c.isalpha() for c in s)
special=len(s)-spaces-digits-alpha
print(spaces,digits,alpha,special)
```

33. Find all occurrences of a substring in "This is Python and Python is fun" → Substring "Python".

```python
s="This is Python and Python is fun"
sub="Python"
start=0
pos=[]
while True:
    i=s.find(sub,start)
    if i==-1:
        break
    pos.append(i)
    start=i+1
print(pos)
```

45. Find the longest common substring between two strings (e.g., "abcdxyz" and "xyzabcd" → Longest common substring = "abcd".

```python
a="abcdxyz"
b="xyzabcd"
longest=""
for i in range(len(a)):
    for j in range(i+1,len(a)+1):
        sub=a[i:j]
        if sub in b and len(sub)>len(longest):
            longest=sub
print(longest)
```
