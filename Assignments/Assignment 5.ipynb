## Assignment-5: String in Python


```python
#1. Filter vowels and consonants from the string "How are you sir".

str1=input("Enter Your string:- ")
vowels=""
consonents=""
specials=""
for i in str1:
  if i in "aeiouAEIOU":
    vowels+=i
  elif i  in "bcdfghjklmnpqursvwxyzBCDFGHJKLMNPQRSTVWXYZ":
    consonents+=i
  else:
    specials+=i

print("vowels are :- ",vowels)
print("Consonents are :- ",consonents)
print("Special characters are:- ",specials)




```
