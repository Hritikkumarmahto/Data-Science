```python
import random

list1=["rock","paper","seaesor"]

def randplay(rand):
  print("""
Enter your selection
      1.Rock
      2.paper
      3.seasor
      """)

  n=input(" Enter your choice by typing :- ")

  if rand==n:
   print("Tie")
  elif rand=="seasor" and n=="rock":
   print("you won")
  elif rand=="seasor" and n=="paper":
   print("better luck")
  elif rand=="rock" and n=="paper":
    print("You won")
  elif rand=="rock" and n=="seasor":
    print("better luck")
  elif rand=="paper" and n=="rock":
   print("You won")
  elif rand=="paper" and n=="seasor":
    print("better luck")
  else:
    print("go play cricket")

rand=random.choice(list1)
randplay(rand)

print("do you want to play again ")
retry=input("y/n")
if(retry=='y'):
  randplay()
else:
  print("thank you for playing")
```
