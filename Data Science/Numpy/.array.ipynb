Upcasting in NUMPY
- Upcasting in numpy refers automatic convrsion of data types when performing operations involving arrays of diffrent types, so that no data lost and the result has a compatible type that can hold all input values.

  - converting int to float is UPcasting - so no data loss  - if we have some float value present in the array
  - Converting float to int is DOwncasting

Two Dimensional numpy array


```python
np.array([[1,2,3],[4,5,6]])
```

```python
# creating 2D array with ndmin
np.array([1,2,3,4], ndmin=2)   # the list should be minimum 2 dimension if not it will automatically convert 

```

```python
np.array([[[2,3,42,4]]], ndmin=3)    # minimum 3D array 
```

```python
np.array([2,3,4],dytype=complex)   # will convert in to complex real + imaginary
```
