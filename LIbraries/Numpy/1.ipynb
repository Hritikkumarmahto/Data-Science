Record arrray or structured array
1. Structured array in numpy are homogeneous at the array level (each element has the same layout)
2. we want to specify a type and, optionally, a name on a paper per column basis, This make sorting and filtering even more powerful, and it can feel similar to working with data in Excel and CSVs or relationsal databses  

```python
records=[("hritik",12,34),
         ("tanmay",34,23),
         ("abhi",24,122)]
print(records)
```

Converting the string elements in the list as int and keeping string element as string usign some methods


```python
array_records=np.array(records,dtype=[("name","U15"),("age",np.int32),("marks",np.int32)])
array_records
```

We can specify the data type in sevral equivalent ways.
1. Using NUMPY type string ( compact )- [("name","U15"), ("age",np.int32),("marks",np.int32)]
2. Using NUMPY type object (recommended)- [("name",np.unicode_),("age",np.int32),("marks",np.int32)]
3. Mixing python build in types and numpy types - [("name",str),("age",int),("marks",np.int32)]




```python
print(array_records[0])   # print the first record of the dataset
```

```python
print(array_records["name"])          # will print the names of the records in the array 
```

Homogeneous array allow:
1. Vecotrized Operations
2. Effictient memeory layout (continuous block)
3. Fast computation using C under the hood

- All elements in numpy array share the same data type (dtype)
