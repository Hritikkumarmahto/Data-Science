# logspace
- np.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=none)
    - start → Starting exponent
    - stop → Ending exponent
    - num → Number of values to generate
    - base → Base of the logarithm (default is 10)

```python
# np.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=none)
arr=np.logspace(1,3,3)
print(arr)
```

```python
arr=np.logspace(0,4,5)
print(arr)

# output means
1
10
100
1000
10000
```

```python
np.logspace(1,5,5,base=2)   


# Output:

# [ 2.  4.  8. 16. 32.]

# Because it calculates:

# 2¹ = 2
# 2² = 4
# 2³ = 8
# 2⁴ = 16
# 2⁵ = 32
```

```python
# it is useful when your data spans many orders of magnitude, such as 
# - signal processing 
# - physics
# - Machine learning 
# - sciencetifuc computation
# - Frequency analysis 

# for example when testing learning rates in machine learning 
arr=np.logspace(-4,0,5)
print(arr)

# Output:

# [0.0001 0.001 0.01 0.1 1. ]

# Instead of checking every value between 0.0001 and 1, you test values that increase by powers of 10, which is much more practical.
```

# Diagonal . Triangular , Lower Triangular , and Upper Triangular Matrix

```python
# to extract diagonal elements 
# Diagonal - Meanig where i = j 
arr=np.array([[1,2,3],[5,6,7],[8,9,10]])

print("Diagonal Elements are",np.diag(arr))
```

```python
# To create a diagonal matrix using diagonal elements for this arrau
np.diag(np.diag(arr))   # using np.diag 2 times one for extracting the diagonal elements another one is for trnasforming 
```

```python
# Lower Triangular matrix

np.tril(arr)  # tri = Triangular , l = lower
```

```python
# Upper Triangular matrix

np.triu(arr)   # tri= Triangular , u = upper
```
